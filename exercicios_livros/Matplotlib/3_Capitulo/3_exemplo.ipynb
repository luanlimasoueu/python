```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('bug.jpg')
print(img)

plt.imshow(img)
```

```python
lum_img = img[:, :, 0]
plt.imshow(lum_img)
```

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

```python
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(lum_img)
a.set_title('Before')
plt.colorbar( orientation='horizontal')
b = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img)
b.set_title('After')
plt.colorbar(orientation='horizontal')
```
