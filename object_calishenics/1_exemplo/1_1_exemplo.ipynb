🏋️ As 9 regras (com exemplos simples)

1️⃣ Apenas um nível de indentação por método

❌ errado

```python
def processar(pedido):
    if pedido:
        if pedido.valido:
            for item in pedido.itens:
                print(item)
```

Muitos níveis = código difícil.

✅ correto

```python
def processar(pedido):
    if not pedido or not pedido.valido:
        return
    mostrar_itens(pedido)

def mostrar_itens(pedido):
    for item in pedido.itens:
        print(item)
```

👉 Quebre em métodos menores.

2️⃣ Não use "else"

Prefira retorno antecipado.

❌

```python
def desconto(cliente):
    if cliente.vip:
        return 0.2
    else:
        return 0.1
```

✅

```python
def desconto(cliente):
    if cliente.vip:
        return 0.2
    return 0.1
```

Código fica mais linear.

3️⃣ Encapsule primitivos

Evite usar tipos básicos diretamente.

❌

```python
saldo = 1000
```


✅

```python
class Saldo:
    def __init__(self, valor):
        self.valor = valor
```

Benefícios:

Validação

Regras de negócio

Segurança

4️⃣ Coleções devem estar encapsuladas

Nunca exponha listas diretamente.

❌

```python
class Carrinho:
    def __init__(self):
        self.itens = []
```

✅

```python
class Carrinho:
    def __init__(self):
        self._itens = []

    def adicionar(self, item):
        self._itens.append(item)
```
