```python
def log(func):
    def wrapper(*args, **kwargs):
        print("Argumentos:", args, kwargs)
        return func(*args, **kwargs)
    return wrapper


@log
def soma(a, b):
    return a + b

soma(10, 20)
```

```python
def log(func):
    def wrapper(*args, **kwargs):
        print("KWARGS:",args)
        print("KWARGS:", kwargs)
        return func(*args, **kwargs)
    return wrapper


@log
def cadastrar(numero,nome, idade):
    print("Cadastrado:", numero, nome, idade)


cadastrar(1, nome="João", idade=30)
```

```python
def debug(func):
    def wrapper(*args, **kwargs):
        print("ARGS:", args)
        print("KWARGS:", kwargs)
        return func(*args, **kwargs)
    return wrapper


@debug
def teste(a, b, nome=""):
    print(a, b, nome)


teste(10, 20, nome="Luan")
```
